```python
x = [i for i in range(5)]
```

```python
plt.plot(x,y, label='Marked Line', color='#ff00ff', linewidth=2, marker='*', markeredgecolor='blue', 
         markersize=10, linestyle='--')
plt.title('First Graph For Analysis', fontdict={'fontname': 'Franklin Gothic Medium', 'fontsize': 20, 
                                                'color': 'orange', 'verticalalignment': 'baseline'}, 
          loc='center')
plt.xlabel('Cost per Goods')
plt.ylabel('Price in USD')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,4), dpi=100)
plt.plot(x,y, 'b*--', label='2*x', markeredgecolor='red')
plt.title('The Graph', fontdict={'color': 'purple', 'fontname': 'Comic sans MS', 'fontsize': 14})
plt.xlabel('Cost')
plt.ylabel('Price')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```

## Bar Chart

```python
labels = ['2018', '2019', '2020', '2021', '2022']
values = [34, 79,90,21,53]
```

```python
plt.figure(figsize=(7,5), dpi=100)
plt.bar(labels, values, label='Population in Millions', color='red')
plt.title('Population Growth', fontdict={'fontname': 'Comic sans MS', 'fontsize':20, 'color':'blue'})
plt.xlabel('Year', fontdict={'color': 'purple'})
plt.ylabel('Population', fontdict={'fontname': 'Franklin Gothic Medium', 'color': 'orange',
                                  'fontsize':12})
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,5), dpi=100)
plt.barh(labels, values, label='Population in Millions', color='red')
plt.title('Population Growth', fontdict={'fontname': 'Comic sans MS', 'fontsize':20, 'color':'blue'})
plt.xlabel('Population', fontdict={'color': 'purple'}, loc='right')
plt.ylabel('Year', fontdict={'fontname': 'Franklin Gothic Medium', 'color': 'orange',
                                  'fontsize':12})
plt.legend()
plt.show()
```

```python
X = ['Group A', 'Group B', 'Group C', 'Group D']
Ygirls = [10,20,20,40]
Zboys = [20,30,25,30]
X_axis = np.arange(len(X))
```

```python
plt.figure(figsize=(8,5), dpi=100)
plt.bar(X_axis-0.2, Ygirls, 0.4, label='Girls')
plt.bar(X_axis+0.2, Zboys, 0.4, label='Boys')
plt.xticks(X_axis, X)
plt.yticks([10,20,30,40,50])
plt.xlabel('Groups')
plt.ylabel('Number of Students')
plt.title(label='Number of Students in each group')
plt.legend()
plt.show()
```

# Pie Chart

```python
x = np.array([30,20,10,15,5,20])
labels = ['Lagos', 'Ogun', 'Anambra', 'Oyo', 'Kano', 'Delta']
explodes = (0.1,0.05,0.1,0.05,0.1,0.05)
colors = ('orange', 'cyan', 'brown', 'grey', 'indigo', 'beige')
```

```python
plt.figure(figsize=(10,7), dpi=150)
plt.pie(x, labels=labels, startangle=90, explode=explodes, colors=colors)
plt.show()
```

```python
plt.figure(figsize=(10,7), dpi=50)
plt.pie(x, labels=labels, startangle=90, explode=explodes, colors=colors, autopct='%.2f%%', pctdistance=0.6)
plt.show()
```

## Subplot

```python
x = np.array([0,1,2,3,4])
y = np.array([15,8,10,15,3])
```

```python
plt.figure(figsize=(10,7), dpi=100)

plt.subplot(2,3,1)
plt.title('Monthly Expenses', loc='center', fontdict={'fontsize': 10, 'color': 'brown'}, pad=20)
plot1 = plt.plot(x,y, color='red', label='First')

plt.subplot(2,3,2)
plt.title('Monthly Expenses', loc='center', fontdict={'fontsize': 10, 'color': 'brown'}, pad=20)
plot2 = plt.plot(x,y, color='green')

plt.subplot(1,3,3)
plt.plot(x,y, color='blue')

plt.subplot(2,3,4)
plt.plot(x,y, color='purple', label='Fourth')
plt.legend()

plt.subplot(2,3,5)
plot5 = plt.plot(x,y, color='yellow', label='2x')
plt.legend()

# plt.legend([plot1, plot2], ['1st', '2nd'], title='Price', loc='upper right')

plt.suptitle('Everything', horizontalalignment='center', fontdict={'color': 'indigo', 'weight': 'bold'})
plt.show()
```

```python
a = np.linspace(0,10,11)
b = a**4 
x = np.arange(0,10)
y = 2**x
```

```python
fig = plt.figure(figsize=(8,5), dpi=200)
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
fig.savefig('new-figure.png', bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y, color='red')
axes[1].plot(a,b, color='green');
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(a,b, color='green')
axes[0][1].plot(x,y, color='red')
axes[1][0].plot(a,b, color='yellow')
axes[1][0].set_title("Title")

axes[1][1].plot(x,y, color='indigo')

fig.subplots_adjust(wspace=1, hspace=1)
fig.suptitle('Figure Level')

fig.set_figwidth(10)
fig.set_figheight(10)
# plt.tight_layout()

plt.show()
```
